```python
'''
Code for computing Fisher-Information matrix using known formula

Problem 3.1 e

'''

#Computes dot product of two vectors
def info_dot_product(theta_hat, x):
     return np.dot(np.transpose(theta_hat), np.transpose(x))

#converts array with one-element to scalar
def turn_scalar(dot_pr):
    return dot_pr[0]

#Compute exponent
def expo(scalar):
    return np.exp(-scalar)

#Compute scalar coefficient (e ^ -theta^T xi)/ (1 + e ^ -theta^T xi)^2
def compute_coeff(scalar):
    exp = expo(scalar)
    num = exp
    denom = (1+ exp)**2
 
    return num/denom

# create matrix by multiplying xi xi^T
def generate_matrix(x):
    arr = np.multiply(x, np.transpose(x))
    return arr

#scalar multiplication of matrix with coefficient
def multiply_coeff_mat(coeff, matrix):
    return np.multiply(coeff, matrix)

#Put everything together to compute Fisher-Information Matrix
def create_fisher_information(theta_hat, X):
    
    n = X.shape[0]
    fisher_matrix = np.zeros((7,7))
    
    for i in range(n):
        
        dot_p = info_dot_product(theta_hat, X[i])
        scalar = turn_scalar(dot_p)
        scalar_coeff = compute_coeff(scalar)
        matx = generate_matrix(X[i][np.newaxis])
        matx = multiply_coeff_mat(scalar_coeff, matx)
        fisher_matrix = np.add(fisher_matrix, matx)
    
    return fisher_matrix


#Generate Fisher-Information Matrix
save_fisher_info_matrix = create_fisher_information(save_theta, X)

#Compute Inverse
inverse = np.linalg.inv(save_fisher_info_matrix)
```
